```python
#list:ordered collection of items
numbers = [10,20,30,40]
#dictionary: collection of key-value pairs
user ={"id":1,"name":"Ethan","is_active":True}
print("numbers", numbers)
print("user", user)
```

```python
#list comprehension
#add 5 to each number in the list
plus_five = [x + 5 for x in numbers]
# filter numbers >20 
greater_than_20 = [n for n in numbers if n > 20]
print("plus_five", plus_five)
print("greater_than_20", greater_than_20)
```

```python
#simple function
def normalize_name(name:str) -> str:
    """Lowercase and strip spaces"""
    return name.strip().lower()

print(normalize_name("  Ethan  "))
```

```python
import json
from typing import List, Dict

sample_users = [
    {"id": 1, "name": "  Alice  ", "active": True},
    {"id": 2, "name": "Bob", "active": False},
    {"id": 3, "name": "  CHARLIE ", "active": True}
    ]

def transform_users(users: List[Dict]) -> List[Dict]:
    """
       -keep only active users 
       - trim & lowercase names
       """
    result = []
    for u in users:
        if not u.get("active"):
            continue
        cleaned = {
            "id": u["id"],
            "name": u["name"].strip().lower()
        }
        result.append(cleaned)
    return result

transformed = transform_users(sample_users)
transformed

```
